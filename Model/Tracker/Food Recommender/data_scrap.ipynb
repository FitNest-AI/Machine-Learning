```python

from bs4 import BeautifulSoup
import requests

def get(food_name):

    site = requests.get("https://www.fatsecret.co.id/kalori-gizi/umum/" + food_name.replace(" ", "-"))
    bsoup = BeautifulSoup(site.text, features="html.parser")

    facts = bsoup.find_all('td', class_='fact')

    data = ['Calories', 'Fat', 'Carbs', 'Protein']
    values = []
    for fact in facts:
        value = fact.find('div', class_='factValue').text
        values.append(float(value.replace('g','').replace('kcal','').strip().replace(',','.')))
    
    return data, values

```

```python
def get_food_group(group_name):
    site = requests.get("https://www.fatsecret.co.id/kalori-gizi/grup/" + group_name)
    bsoup = BeautifulSoup(site.text, features="html.parser")
    food_name = bsoup.find_all("div", class_="food_links")
    food_list = []
    for name in food_name:
        name = name.text.split(',')[0]
        food = get(name)
        print(name,food)
        print("=====================================")
        food_list.append((name,food))
    return food_list
```

```python
food_list = ["biji-bijian","buah","daging","ikan-seafood","kacang-kacangan","keju%C2%A0-susu","makanan-cepat-saji"
             ,"makanan penutup","makanan ringan","nasi%C2%A0mie%C2%A0-%C2%A0pasta","roti-sereal",
             "salad","sayuran","sup",""]

food_data = []

for group in food_list:
    data = get_food_group(group)
    print(data)
    food_data.append(data)
```

```python
food_name = []
label = ['Calories', 'Fat', 'Carbs', 'Protein']
data = []

for food in food_data:
    try:
        print(food[0])
        for i in food:
            food_name.append(i[0])
            data.append(i[1][1])
    except:
        pass
    
df = pd.DataFrame(data, columns=label, index=food_name)
df.dropna(inplace=True)
df
df.to_csv("foods.csv")
```
